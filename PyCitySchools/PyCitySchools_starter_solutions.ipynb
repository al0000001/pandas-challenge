```python
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "./Resources/schools_complete.csv"
student_data_to_load = "./Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Local Government Area (LGA) Summary
total_schools= len(school_data_complete ["school_name"].unique())
total_students= len(school_data_complete ["Student ID"])
total_budget=school_data["budget"].sum()
Average_Maths_Score=school_data_complete ["maths_score"].mean()
Average_Reading_Score=school_data_complete ["reading_score"].mean()
#Percentage Calculation
maths_pass=len(school_data_complete[school_data_complete["maths_score"]>=50])
reading_pass=len(school_data_complete[school_data_complete["reading_score"]>=50])
overall_pass=len(school_data_complete[(school_data_complete["reading_score"]>=50) & (school_data_complete["maths_score"]>=50)])
Percentage_Passing_Maths = maths_pass/total_students*100
Percentage_Passing_Reading	= reading_pass/total_students*100
Percentage_Overall_Passing= overall_pass/total_students*100
summary_df = pd.DataFrame(
    {"Total Schools": [total_schools],
     "Total Students": [total_students],
     "Total Budget": [total_budget],
     "Average Maths Score": [Average_Maths_Score],
     "Average Reading Score": [Average_Reading_Score],
     "% Passing Maths": [Percentage_Passing_Maths],
     "% Passing Reading": [Percentage_Passing_Reading],
     "% Overall Passing": [Percentage_Overall_Passing],
     }
)
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,.2f}".format)
summary_df




```

```python
#School Summary
#----------------------------------------------------IMPORTANT TO READ--------------------------------------------------------------- 
# School name & School type method copied from below link, learnt set_index then add specific colomn by [""] and .groupby can link with .count
#https://notebooks.githubusercontent.com/view/ipynb?browser=chrome&color_mode=auto&commit=7189be806a9ac33a129b6d8476f07d84946666d0&device=unknown&enc_url=68747470733a2f2f7261772e67697468756275736572636f6e74656e742e636f6d2f65726d69617367656c6179652f70616e6461732d6368616c6c656e67652f373138396265383036613961633333613132396236643834373666303764383439343636363664302f5079436974795363686f6f6c732f2e6970796e625f636865636b706f696e74732f5079436974795363686f6f6c735f737461727465722d636865636b706f696e742e6970796e62&logged_in=false&nwo=ermiasgelaye%2Fpandas-challenge&path=PyCitySchools%2F.ipynb_checkpoints%2FPyCitySchools_starter-checkpoint.ipynb&platform=android&repository_id=269505066&repository_type=Repository&version=96
school_name_group = school_data_complete.set_index('school_name').groupby(['school_name'])
school_type = school_data.set_index('school_name')["type"]
total_students_s = school_name_group["Student ID"].count()
total_School_Budget=school_data.set_index('school_name')["budget"]
Per_student_budget=total_School_Budget/total_students_s 
Average_maths_score=school_name_group["maths_score"].mean()
Average_reading_score=school_name_group["reading_score"].mean()
#Percentage Calculation
number_passing_maths=(school_data_complete[school_data_complete["maths_score"]>=50]).groupby("school_name")["Student ID"].count()
number_passing_reading=(school_data_complete[school_data_complete["reading_score"]>=50]).groupby("school_name")["Student ID"].count()
number_passing_overall=school_data_complete[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)].groupby("school_name")["Student ID"].count() 
percentage_passing_Maths=number_passing_maths/total_students_s*100
percentage_passing_reading=number_passing_reading/total_students_s*100
percentage_passing_overall=number_passing_overall/total_students_s*100

school_summary_df = pd.DataFrame({"School Type": school_type,
                                  "Total students": total_students_s,
                                  "Total School budget": total_School_Budget,
                                  "Per student budget": Per_student_budget,
                                  "Average maths score" : Average_maths_score,
                                  "Average reading score" : Average_reading_score,
                                  "% passing maths" : percentage_passing_Maths,
                                  "% passing reading" : percentage_passing_reading,
                                  "% overall passing":percentage_passing_overall,
                                 })

school_summary_df["Total School budget"] = school_summary_df["Total School budget"].map("${:,.2f}".format)
school_summary_df["Per student budget"] = school_summary_df["Per student budget"].map("${:,.2f}".format)
school_summary_df["% passing maths"] = school_summary_df["% passing maths"].map("{:.6f}".format)
school_summary_df["% passing reading"] = school_summary_df["% passing reading"].map("{:.6f}".format)
school_summary_df["% overall passing"] = school_summary_df["% overall passing"].map("{:.6f}".format)
school_summary_df.head(15)

```

```python
#Highest-Performing Schools (by % Overall Passing)
school_summary_df = school_summary_df.sort_values("% overall passing", ascending=False) 
school_summary_df.head(5)
```

```python
#Lowest-Performing Schools (by % Overall Passing)
school_summary_df = school_summary_df.sort_values("% overall passing") 
school_summary_df.head(5)
```

```python
#Maths Scores by Year

grouped_year9 = student_data.loc[student_data['year'] == 9]
grouped_year9_school_math_mean = grouped_year9.groupby("school_name")["maths_score"].mean()

grouped_year10 = student_data.loc[student_data['year'] == 10]
grouped_year10_school_math_mean = grouped_year9.groupby("school_name")["maths_score"].mean()

grouped_year11 = student_data.loc[student_data['year'] == 11]
grouped_year11_school_math_mean = grouped_year11.groupby("school_name")["maths_score"].mean()

grouped_year12 = student_data.loc[student_data['year'] == 12]
grouped_year12_school_math_mean = grouped_year12.groupby("school_name")["maths_score"].mean()

math_summary = pd.DataFrame({"Year 9":grouped_year9_school_math_mean,
                             "Year 10":grouped_year10_school_math_mean,
                            "Year 11":grouped_year11_school_math_mean,
                            "Year 12":grouped_year12_school_math_mean,  
                            })
math_summary
```

```python
#Reading Scores by Year

grouped_year9 = student_data.loc[student_data['year'] == 9]
grouped_year9_school_reading_mean = grouped_year9.groupby("school_name")["reading_score"].mean()

grouped_year10 = student_data.loc[student_data['year'] == 10]
grouped_year10_school_reading_mean = grouped_year9.groupby("school_name")["reading_score"].mean()

grouped_year11 = student_data.loc[student_data['year'] == 11]
grouped_year11_school_reading_mean = grouped_year11.groupby("school_name")["reading_score"].mean()

grouped_year12 = student_data.loc[student_data['year'] == 12]
grouped_year12_school_reading_mean = grouped_year12.groupby("school_name")["reading_score"].mean()

reading_summary = pd.DataFrame({"Year 9":grouped_year9_school_reading_mean,
                             "Year 10":grouped_year10_school_reading_mean,
                            "Year 11":grouped_year11_school_reading_mean,
                            "Year 12":grouped_year12_school_reading_mean,  
                            })
reading_summary
```

```python
#Scores by School Spending
bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
school_data_complete["Spending Ranges (Per Student)"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], bins, labels=spending_labels, include_lowest=True)

spending_maths_scores = school_data_complete.groupby(["Spending Ranges (Per Student)"]).mean()["maths_score"]
spending_reading_scores = school_data_complete.groupby(["Spending Ranges (Per Student)"]).mean()["reading_score"]

number_pass_math = school_data_complete[school_data_complete['maths_score'] >= 50].groupby("Spending Ranges (Per Student)")['Student ID'].count()
number_reading_math = school_data_complete[school_data_complete['reading_score'] >= 50].groupby("Spending Ranges (Per Student)")['Student ID'].count()
number_overall=school_data_complete[(school_data_complete['maths_score'] >= 50) & (school_data_complete['reading_score'] >= 50)].groupby("Spending Ranges (Per Student)")['Student ID'].count()

percentage_pass_math = number_pass_math/school_data_complete.groupby("Spending Ranges (Per Student)")['Student ID'].count()*100
percentage_pass_reading = number_reading_math/school_data_complete.groupby("Spending Ranges (Per Student)")['Student ID'].count()*100
percentage_pass_overall = number_overall/school_data_complete.groupby("Spending Ranges (Per Student)")['Student ID'].count()*100


scores_by_spending_summary = pd.DataFrame({"Averge Maths Score":spending_maths_scores,
                                           "Averge Reading Score":spending_reading_scores,
                                           "% Passing Maths":percentage_pass_math,
                                           "% Passing Reading":percentage_pass_reading,
                                           "% Overall Passing":percentage_pass_overall,
                           
                                         })
scores_by_spending_summary.head()                          


```

```python
#Scores by School Size

size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_data_complete["School Size"] = pd.cut(school_data_complete["size"], size_bins, labels=size_labels, include_lowest=True)


size_maths_scores = school_data_complete.groupby(["School Size"]).mean()["maths_score"]
size_reading_scores = school_data_complete.groupby(["School Size"]).mean()["reading_score"]

number_pass_math = school_data_complete[school_data_complete['maths_score'] >= 50].groupby("School Size")['Student ID'].count()
number_reading_math = school_data_complete[school_data_complete['reading_score'] >= 50].groupby("School Size")['Student ID'].count()
number_overall=school_data_complete[(school_data_complete['maths_score'] >= 50) & (school_data_complete['reading_score'] >= 50)].groupby("School Size")['Student ID'].count()

percentage_pass_math = number_pass_math/school_data_complete.groupby("School Size")['Student ID'].count()*100
percentage_pass_reading = number_reading_math/school_data_complete.groupby("School Size")['Student ID'].count()*100
percentage_pass_overall = number_overall/school_data_complete.groupby("School Size")['Student ID'].count()*100


scores_by_size_summary = pd.DataFrame({"Averge Maths Score":size_maths_scores,
                                           "Averge Reading Score":size_reading_scores,
                                           "% Passing Maths":percentage_pass_math,
                                           "% Passing Reading":percentage_pass_reading,
                                           "% Overall Passing":percentage_pass_overall,
                           
                                         })
scores_by_size_summary.head()       

```
